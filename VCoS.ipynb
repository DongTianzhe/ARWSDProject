## Calculate VCoS

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
import pottslab as pl
import os
```

Read Images

```python
imageList = []
path = 'pic/'

nameList = os.listdir(path)
for name in nameList:
    if name.endswith('.png'):
        imageList.append(cv.imread(path + name))

# OpenCV read image using BGR
imgrgb = cv.cvtColor(imageList[0], cv.COLOR_BGR2RGB)
plt.axis('off')
plt.imshow(imgrgb)
plt.show()
```
