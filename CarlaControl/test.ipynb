```python
import carla
import random
import numpy as np
import cv2 as cv
import time
```

```python
client = carla.Client('localhost', 2000)
world = client.get_world()

bpLib = world.get_blueprint_library()
spawnPoints = world.get_map().get_spawn_points()
```

```python
barrierBP = bpLib.find('static.prop.streetbarrier')
spawnSpot = [carla.Transform(carla.Location(x=41.85, y=294.09, z=0), carla.Rotation()),
             carla.Transform(carla.Location(x=45.85, y=294.09, z=0), carla.Rotation()),
             carla.Transform(carla.Location(x=180.89, y=237.01, z=0), carla.Rotation(yaw=90)),
             carla.Transform(carla.Location(x=180.89, y=241.01, z=0), carla.Rotation(yaw=90)),
             carla.Transform(carla.Location(x=180.89, y=187.61, z=0), carla.Rotation(yaw=90)),
             carla.Transform(carla.Location(x=180.89, y=191.61, z=0), carla.Rotation(yaw=90)),
             carla.Transform(carla.Location(x=5.24, y=187.68, z=0), carla.Rotation(yaw=-90)),
             carla.Transform(carla.Location(x=5.24, y=191.68, z=0), carla.Rotation(yaw=-90))
             ]
print(spawnSpot)
barriers = []
for  i in range(len(spawnSpot)):
    barrier = world.spawn_actor(barrierBP, spawnSpot[i])
    barriers.append(barrier)
```

```python
for i in range(len(spawnSpot)):
    barriers[i].destroy()
```
